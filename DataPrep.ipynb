```python
import yfinance as yf
import pandas as pd
import yfinance as HistoricalPrices
import hvplot.pandas
from pathlib import Path
```

```python
# Whats your ticker ?
ticker = 'ARKK'
# ticker = int(ticker)
# Where data starts and ends ?
start = '2020-03-01'
end = '2022-03-31'
arkk_etf = yf.download(ticker, start, end)
```

```python
# Whats your ticker ?
ticker = 'SPY'
# ticker = int(ticker)
# Where data starts and ends ?
start = '2020-03-01'
end = '2022-03-31'
spy_etf = yf.download(ticker, start, end)
```

```python
# Whats your ticker ?
ticker = 'FNGU'
# ticker = int(ticker)
# Where data starts and ends ?
start = '2020-03-01'
end = '2022-03-31'
fngu_etf = yf.download(ticker, start, end)
```

```python
arkk_etf = arkk_etf["Adj Close"].apply(lambda x: round(x,2)).rename("ARKK")
spy_etf = spy_etf["Adj Close"].apply(lambda x: round(x,2)).rename("SPY")
fngu_etf = fngu_etf["Adj Close"].apply(lambda x: round(x,2)).rename("FNGU")
```

```python
etf_data = pd.concat([arkk_etf, spy_etf, fngu_etf], join="inner", axis = 1)
```
